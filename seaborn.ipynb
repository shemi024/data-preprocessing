```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#load abuilt in dataset based on us state car crash percentage
crash_df = sns.load_dataset('car_crashes')

#distribution plot
#provides a way to look at a uninvarte distribution .A
#univeriate distribution provides a distribution for a one variable 
sns.displot(crash_df['not_distracted'],kde=False, bins=25)
```

```python
#join plot
#joining compares 2 distribution and plot a scatter plot by default
# as we can see as people tend to drink and drive
sns.jointplot(x='speeding',y='alcohol',data=crash_df,kind='reg')
```

```python
#KDE plot
#a kernal density estimate (KDE) plot is a method for visualizing the distribution of observation in a dataset
# get just the KDE plot
sns.kdeplot(crash_df['alcohol'])
```

```python
#pair plots
# pair plot plots relationship across the entire data frames numerical values 
sns.pairplot(crash_df)
tips_df = sns.load_dataset('tips')
```

```python
#plots a single column of dataponts in array as sticks on an axis
#with a rug plot you'll see a more dens numbers of lines where the amount is
#most common . this is like how histogram is taller where values are more common
sns.rugplot(tips_df['tip'])
```

```python
#styling
#you can set styling for your axes and grids
# white,darkgrid,whitegrid,dark,ticks
sns.set_style('white')
plt.figure(figsize=(8,4))
sns.set_context('paper',font_scale=1.4)
sns.jointplot(x='speeding',y='alcohol',data=crash_df,kind='reg')

sns.despine(left=False,bottom=False)
```

```python
#ctegorical plots
#bar plots
#focus on distribution using categorical data in reference to one of the numerical
#columns
#aggregate categorical data based on a function (mean is default)
#estimate total bill amount based on sex
#with estimsate you can define function to use other than the mean like those
#providing by numoy:median,std,var,cov or make your own function
sns.barplot(x='sex',y='total_bill',data=tips_df,estimator=np.median)
```

```python
#boxplot
plt.figure(figsize=(14,9))
sns.set_style('darkgrid')
sns.boxplot(x='day',y='total_bill',data=tips_df,hue='sex')
plt.legend(loc=0)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
 
#violinplot
#it is a combination of boxplot and KDE
#split allows you to compare how the categories compare to each other 
sns.violinplot(x='day',y='total_bill',data=tips_df, hue='sex',split=True)
```

```python
#strip plot
plt.figure(figsize=(8,5))
#the strip plot draws a sctter plot  reprasenting all the data points wher one variable is catogerical 
#it is often used to show all observation with a box that reprasents the avg distribution 
#jiter spreads data ponts out so that they aren't stacked top of each other
# hue breaks data into men and women
#dodge seperates data into men and women data
sns.stripplot(x='day' ,y='total_bill',data=tips_df, hue='sex',jitter=True,dodge=True)
```

```python
#swarm plot
# it is like a strip plot , but points are adjusted so they don't overlap
# it looks like a combination of the violin and strip plots
# sns.swarmplot(x='day',y='total_bill',data=tips_df)
# u can stack violin plot with a swarm

sns.violinplot(x='day',y='total_bill',data=tips_df)
sns.swarmplot(x='day',y='total_bill',data=tips_df, color='white')
```

```python
# pallettes
plt.figure(figsize=(8,6))
sns.set_style('dark')
sns.set_context('talk')

# u can use matplotlips color maps for color styling
# https://matplotlib.org/3.3.1/tutorials/colors/colormaps.html
sns.stripplot(x='day',y='total_bill',data=tips_df,hue='sex',palette='seismic')
# add the optical legend with a location no(best:0,upper right:1,upper left;2,lower left right:4,)
# or supply a tuple  of x & y from lower left
plt.legend(loc=0)
```
